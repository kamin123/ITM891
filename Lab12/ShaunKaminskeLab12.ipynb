### Problem 1 - Basic Features of Graphs
Here we will create a complete graph of 5 vertices and determine basic features about the graph, such as the number of edges it has, the degree, etc. Refer to the documentation here for help: https://networkx.org/documentation/stable/tutorial.html.



```python
import networkx as nx

# Enter your answer under each step.

# Create a complete graph of 5 vertices
G = nx.Graph()
for i in range(5):
  G.add_node(i)
G.add_edge(0,1)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(4,3)
G.add_edge(4,0)
G.add_edge(4,2)
G.add_edge(1,4)

# Plot the graph
import matplotlib.pyplot as plt
nx.draw(G, with_labels=True, font_weight='bold')

# How many edges in this graph?
edges = G.number_of_edges()

#7 edges

# What is the adjacency matrix for this graph?
adj_matrix = G.adj
# {0: {1: {}, 4: {}}, 1: {0: {}, 2: {}, 4: {}}, 2: {1: {}, 3: {}, 4: {}}, 3: {2: {}, 4: {}}, 4: {3: {}, 0: {}, 2: {}, 1: {}}}

# What is the degree for every vertex?
degrees = dict(G.degree)
# {0: 2, 1: 3, 2: 3, 3: 2, 4: 4}

# What are the neighbors of the vertex 0 in this graph?
neighbors = list(G.adj[0])
#[1, 4]
```

```python
# Calculate the shortest path between any two vertices (all the pairs in the graph).
sp = dict(nx.all_pairs_shortest_path(G))
spl = dict(nx.all_pairs_shortest_path_length(G))


print(f'The shortest path from vertex 0 to vertex 2 is {sp[0][2]} and the length of this shortest path is {spl[0][2]}')


```

```python
# Calculate the distance between each pair of vertices (0 and 1, 1 and 2, 2 and 3, etc.).
max_distance = 2
max_path = []
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance == max_distance:
      max_path.append((source_node, target_node))
for source, target in max_path:
  print(len(nx.shortest_path(G, source, target)))
```

```python
# Calculate the diameter of the graph.
list(nx.shortest_path_length(G))
max_distance = -1
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance > max_distance:
      max_distance = distance
print(f"max distance: {max_distance}")
```

```python
# Is the graph connected? Check to see if it is. 
connections = list(nx.connected_components(G))
print(connections)
#yes it is connected
```

```python
# How many connected components are in the graph?
nx.number_connected_components(G)
```

```python
# What is the path with the distance equal to this diameter?
# Hint: Longest distance, nx.shortest_path(G)


list(nx.shortest_path_length(G))
max_distance = -1
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance > max_distance:
      max_distance = distance


# Find the pair of nodes with the longest distance
# max_distance = -1
max_pair = None
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance == max_distance:
      max_pair = (source_node, target_node)

print(f'Pair of Nodes with longest distance: {max_pair}')
# Get the path corresponding to the longest distance
print(f'Path corresponding to longest distance: {nx.shortest_path(G, max_pair[0], max_pair[1])}')
```
